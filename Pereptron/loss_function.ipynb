```python
from sklearn.datasets import make_classification

X,y = make_classification(n_samples=100,n_features=2,n_informative=1,n_redundant=0,random_state=41,hypercube=False,n_classes=2,class_sep=20,n_clusters_per_class=1)
```

```python
y[y==0] = -1
```

```python
def perceptron(X,y):
    w1=w2=b=1
    lr = 0.1
    epochs=1000
    for _ in range(epochs):
        for i in range(X.shape[0]):
            z = w1*X[i,0]+w2*X[i,1]+b
            if z*y[i] < 0:
                w1 = w1 + lr*y[i]*X[i,0]
                w2 = w2 + lr*y[i]*X[i,1]
                b = b + lr*y[i]
    return b,w1,w2
```

```python
b,w1,w2 = perceptron(X,y)
```

```python
m = -w1/w2
c = -b/w2
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter')
plt.ylim(X[:,1].min()-1,X[:,1].max()+1)
```

```python
line_x = np.linspace(X[:,0].min()-1,X[:,0].max()+1)
line_y = m*line_x+c
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter')
plt.plot(line_x,line_y,color='red')
plt.ylim(X[:,1].min()-1,X[:,1].max()+1)
```
