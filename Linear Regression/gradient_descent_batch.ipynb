```python
from sklearn.datasets import load_diabetes
import numpy as np
```

```python
X,y = load_diabetes(return_X_y=True)
y.shape
```

```python
class GDRegressor:
    def __init__(self,lr=0.1,epochs=1000):
        self.lr = lr
        self.epochs =epochs
        self.coef_ = None
        self.intercept_ = 0
    def fit(self,X,y):
        self.coef_ = np.ones(X.shape[1])
        for _ in range(self.epochs):
            y_hat = np.dot(X,self.coef_)+self.intercept_
            self.intercept_ = self.intercept_ + 2* self.lr * np.mean(y-y_hat)
            self.coef_ = self.coef_ + 2/X.shape[0] * self.lr * np.dot(y-y_hat,X)
        return self.intercept_,self.coef_
    def predict(self,X):
        return np.dot(X,self.coef_)+self.intercept_
```

```python
reg =GDRegressor()
intercept_,coef_ = reg.fit(X,y)
intercept_
```
